```python
import pandas as pd
```

```python
link = 'https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2025-11.parquet'
```

```python
pq = pd.read_parquet(link)
```

```python
ti = pq['lpep_pickup_datetime'].count()
ti
```

```python
november_trips = pq[(pq['lpep_pickup_datetime'] >= '2025-11-01') & (pq['lpep_pickup_datetime'] < '2025-12-01')]
```

```python
count = len(november_trips)
count
```

```python
# Number of trips between November and December with a trip distance <= 1 mile.
mask = (
    (pq['lpep_pickup_datetime'] >= '2025-11-01') & 
    (pq['lpep_pickup_datetime'] < '2025-12-01') & 
    (pq['trip_distance'] <= 1)
)
```

```python
trip_count = len(pq[mask])
trip_count
```

```python
# Filter for trips less than 100 miles
filtered_pq = pq[pq['trip_distance'] < 100]
```

```python
daily_max = filtered_pq.groupby(filtered_pq['lpep_pickup_datetime'].dt.date)['trip_distance'].max()
```

```python
# Find the day with the overall maximum
longest_trip_day = daily_max.idxmax()
longest_distance = daily_max.max()

print(f"The day with the longest trip was {longest_trip_day} with {longest_distance} miles.")
```

```python
# Filter for only November 18th, 2025
nov_18_data = pq[pq['lpep_pickup_datetime'].dt.date == pd.to_datetime('2025-11-18').date()]

# 3. Group by the Pickup Zone and sum the total_amount
zone_totals = nov_18_data.groupby('PULocationID')['total_amount'].sum()

# 4. Get the zone with the maximum sum
top_zone = zone_totals.idxmax()
top_amount = zone_totals.max()

print(f"Zone ID {top_zone} had the highest total amount: ${top_amount:.2f}")
```
